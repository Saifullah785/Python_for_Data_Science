```python
# scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level functions
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

```python
# relplot -> figure level -> square shape

sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
# axes level function
sns.lineplot(x='year',y='lifeExp',data=temp_df)
plt.show()
```

```python
# using relplot
sns.relplot(x='year',y='lifeExp',data=temp_df,kind= 'line')
plt.show()
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['Pakistan','Brazil','Germany'])]
temp_df.sample(5)
```

```python
sns.relplot(x='year',y='lifeExp',data=temp_df,kind='line',hue='country',style='continent',size='continent')
plt.show()
```

```python
sns.lineplot(x='year',y='lifeExp',data=temp_df,hue='country',style='country')
plt.show()
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='day' , row='sex')
plt.show()
```

```python
## it will not work with sns.scatter plot
# sns.scatterplot(x='total_bill',y='tip',data=tips,col='day' , row='sex')
# plt.show()
```

```python
# col wrap
sns.relplot(data=gap, x= 'lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
plt.show()
```

```python
# lineplot in seaborn
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

# **2.Distribution Plots**

Used to find out the distribution

Range of the observation

Central Tendancy

Is the data bimodal?

Are there outliers?

# **Plots under distribution plot**

histplot

kdeplot

rugplot



```python
# figure level -> distplot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# ploting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill',kind='hist',bins=20)
```

```python
# it's also possible to visulize the distribution of a categorical variable using the logic of a hitogram.

# Discrete bins are automatically set for categorical variables

# countplot

sns.displot(data=tips, x='day',kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip',kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=tips, x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data= titanic,x='age',kind='hist',element='step',hue='survived')
```

```python
# faceting using col and row
sns.displot(data=tips, x='tip',kind='hist',col='sex',element='step')
```

```python
# sns.kdeplot
# rather than using Discret bins, a KDE plot smooths the observation with a Gaussian kernel, producing a continous density estimate
plt
sns.kdeplot(data=tips, x='total_bill')
```

```python
# hue -> fill
# sns.kdeplot(data=tips, x='total_bill',hue='sex',fill=True)

sns.displot(data=tips, x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing tricks along the x and y axes.

# this function is intended to complement other plots by showing the location of individual observation in an unobtrusive way.
sns.kdeplot(data=tips, x='total_bill')
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# Bivariate histogram

# A bivariate histogram bins the data within rectangles that tile the plot

# and than shows the count of observation within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill',y='tip')

# sns.distplot(data=tips, x='total_bill', y='tip', kind='hist')

```

```python
# Bivariate kdeplot
# a bivariate kde plot smothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(data=tips, x='total_bill',y='tip')
```

# **3.Matrix Plot**

Heatmap

Clustermap

```python
# heatmap
# plot rectangular data as color encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')


# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='summer') # cmap # linewidth
```

```python
# Clustermap

# plot a matrix datset as a hierarchically - clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
