```python
# scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level functions
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

```python
# relplot -> figure level -> square shape

sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
# axes level function
sns.lineplot(x='year',y='lifeExp',data=temp_df)
plt.show()
```

```python
# using relplot
sns.relplot(x='year',y='lifeExp',data=temp_df,kind= 'line')
plt.show()
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['Pakistan','Brazil','Germany'])]
temp_df.sample(5)
```

```python
sns.relplot(x='year',y='lifeExp',data=temp_df,kind='line',hue='country',style='continent',size='continent')
plt.show()
```

```python
sns.lineplot(x='year',y='lifeExp',data=temp_df,hue='country',style='country')
plt.show()
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='day' , row='sex')
plt.show()
```

```python
# col wrap
sns.relplot(data=gap, x= 'lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
plt.show()
```

```python
# lineplot in seaborn
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```
