```python
# importing libraries
import seaborn as sns
```

```python
# load the datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")

```

# **Categorical Plots**

**Categorical Scatter Plot**

Stripplot

Swarmplot

**Categorical Distribution Plots**

Boxplot

Violinplot

**Categorical Estimate Plot->for central tendency**

Barplot

Pointplot

Countplot

Figure level function-> catplot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# strip plot
sns.stripplot(data=tips, y="total_bill", x="day") #jitter=False
```

```python
# using catplot
# -> figure level function
sns.catplot(data=tips, y="total_bill", x="day", kind="strip",jitter=0.1,hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, y="total_bill", x="day",kind='swarm')
```

```python
sns.swarmplot(data=tips, y="total_bill", x="day",hue='sex')
```

# **Boxplot**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary("minimun",first quartile Q1, median, third quartile Q3 and 'Maximum'.

it can tell you about your outliers and what their values are.

Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# using catplot
sns.catplot(data=tips, y="total_bill", x="sex", kind="box")
```

```python
# hue
sns.catplot(data=tips, y="total_bill", x="day", kind="box",hue='sex')
```

```python
# single boxplot - > numerical col
sns.boxplot(x=tips['total_bill'])
```

```python
# Violinplot = (Boxplot + KDE PLot)
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# hue
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
# using catplot for violin
sns.catplot(data=tips, y="total_bill", x="day", kind="violin")
```

```python
sns.catplot(data=tips, y="total_bill", x="day", kind="violin",hue='sex',split=True)
```

```python
# barplot
sns.barplot(data=tips, x="sex", y="total_bill",hue='smoker',estimator='min')
```

When there are multiple observation in each category, its also uses bootstapping to compute a confidence interval around the estimate , which is plotted using error bars

```python
sns.catplot(data=tips, x="sex", y="total_bill",kind='box',errorbar=None)
```

```python
sns.pointplot(data=tips, x="sex", y="total_bill",hue='smoker',ci=None)
```

```python
# pointplot
sns.pointplot(data=tips, x="sex", y="total_bill",hue='smoker',estimator='min')
```

```python
# countplot
sns.countplot(data=tips, x="sex",hue='day')
```
