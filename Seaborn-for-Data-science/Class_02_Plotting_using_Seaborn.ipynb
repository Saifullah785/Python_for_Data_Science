**Introduction to Categorical Plots:**

It begins by outlining different types of categorical plots: scatter plots (stripplot, swarmplot), distribution plots (boxplot, violinplot), and estimate plots (barplot, pointplot, countplot), highlighting the use of the figure-level function catplot for flexibility.

**Stripplot and Swarmplot:**

Demonstrates creating basic stripplots and swarmplots, and then uses catplot with hue to add another categorical variable for comparison.

**Boxplot:**

Explains the concept of a boxplot and its five-number summary, showing how to create a standard boxplot and then using catplot with and without hue for faceted boxplots. It also shows a single boxplot for a numerical column.

**Violinplot:**

Introduces the violinplot as a combination of a boxplot and a kernel density estimate (KDE) plot, illustrating its use with and without hue, and showing how to use catplot with split to compare distributions within categories.

**Barplot and Pointplot:**

Covers the barplot for showing central tendency with confidence intervals and the pointplot for comparing the estimate of a central tendency across categories, demonstrating the use of hue and estimator (for pointplot).

**Countplot:**

Shows how to use a countplot to display the number of observations in each category.

**Faceting with Catplot:**

Explores the power of catplot for creating faceted plots based on different categorical variables using col and row.

**Regression Plots:**

Introduces regplot and lmplot for visualizing linear relationships between two variables, including the concept of a regression line and confidence interval, and shows how lmplot allows for faceting with hue.

**Residplot:**

Demonstrates the use of residplot to plot the residuals of a regression model.

**FacetGrid, PairGrid, and JointGrid:**

Shows how to use these lower-level functions to create more customized multi-panel plots, including pairwise relationships with PairGrid (mapping different plot types to diagonal, upper, and lower panels) and joint distributions with JointGrid (combining different plot types for marginal and bivariate views).

```python
# importing libraries
import seaborn as sns
```

```python
# load the datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")

```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# strip plot
sns.stripplot(data=tips, y="total_bill", x="day") #jitter=False
```

```python
# using catplot
# -> figure level function
sns.catplot(data=tips, y="total_bill", x="day", kind="strip",jitter=0.1,hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, y="total_bill", x="day",kind='swarm')
```

```python
sns.swarmplot(data=tips, y="total_bill", x="day",hue='sex')
```

# **Boxplot**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary("minimun",first quartile Q1, median, third quartile Q3 and 'Maximum'.

it can tell you about your outliers and what their values are.

Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# using catplot
sns.catplot(data=tips, y="total_bill", x="sex", kind="box")
```

```python
# hue
sns.catplot(data=tips, y="total_bill", x="day", kind="box",hue='sex')
```

```python
# single boxplot - > numerical col
sns.boxplot(x=tips['total_bill'])
```

```python
# Violinplot = (Boxplot + KDE PLot)
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# hue
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
# using catplot for violin
sns.catplot(data=tips, y="total_bill", x="day", kind="violin")
```

```python
sns.catplot(data=tips, y="total_bill", x="day", kind="violin",hue='sex',split=True)
```

```python
# barplot
sns.barplot(data=tips, x="sex", y="total_bill",hue='smoker',estimator='min')
```

When there are multiple observation in each category, its also uses bootstapping to compute a confidence interval around the estimate , which is plotted using error bars

```python
sns.catplot(data=tips, x="sex", y="total_bill",kind='box',errorbar=None)
```

```python
sns.pointplot(data=tips, x="sex", y="total_bill",hue='smoker',ci=None)
```

```python
# pointplot
sns.pointplot(data=tips, x="sex", y="total_bill",hue='smoker',estimator='min')
```

```python
# countplot
sns.countplot(data=tips, x="sex",hue='day')
```

```python
# pointplot
```

```python
# faceting using catplot
sns.catplot(data=tips, x="sex", y="total_bill",col='smoker',kind='box',row='time')
```

in the simplest invocation, both functions draw a scatterplot of two variables, x and y and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression

# **Reg plot**

```python
# axes level

# hue parameter is not awailable

sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip",hue='sex')
```

```python
# residplot
sns.residplot(data=tips, x="total_bill", y="tip")
```

**A second way to plot Facet plots ->FacetGrid**

```python
# figure level ->relplot ->diplot -> catplot->lmplot
sns.catplot(data=tips, x="sex", y="total_bill",col='day',kind='violin',row='time')
```

```python
g = sns.FacetGrid(data=tips, col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

# **Plotting Pairwise Relationship (pairgrid vs pairplot)**

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag --> map_offdiag
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag --> map_offdiag
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.histplot)
```

```python
# map_diag --> map_upper --> map_lower
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.violinplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# vars
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.violinplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

# **JointGrid Vs JointPlot**

```python
sns.jointplot(data=tips, x="total_bill", y="tip",kind='hist',hue='sex') #'scatter', 'hist', 'hex', 'kde', 'reg', 'resid
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot,sns.histplot)
```
