```python
# looping on groups
df_list = [] # Initialize an empty list to store DataFrames
for group,data in genres:
  # Append the filtered data to the list
  df_list.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

# Concatenate all DataFrames in the list
df = pd.concat(df_list)

df
```

```python
# split (apply) combine
# apply  -> builtin function
genres.apply(min)
```

```python
# find number of movies starting with A for each group
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find rating of each movie in the group according to IMDB score

def rank_movie(group):
  group['genres_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise
def normalize_rating(group):
  group['normalized_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(normalize_rating)
```
