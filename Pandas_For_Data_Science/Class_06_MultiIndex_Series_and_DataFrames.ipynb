# **Long Vs Wide Data**

wide formate is where we have a single row for every data point with multiple columns to hold the values of various attributess.

Long formate is where ,for each data point we have as many rows as the number of attributes and each row contains the value of a particular
attribute for a given data point.

```python
# melt -> simple example branch
# wide to long
pd.DataFrame({'cse':[120]}).melt()
```

```python
# melt -> branch with year
pd.DataFrame({'cse':[120],'ece':[130],'mech':[50]}).melt(var_name='branch',value_name="num_students")
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[120,130,50],
        '2021':[120,130,50],
        '2022':[120,130,50]
    }
).melt(id_vars='branch',var_name='year',value_name='num_students')
```

# **Pivot Table**

The pivot table takes simple column-wise data as input, and groups the entries into a two -dimensional table that provide a multidimensional summarization of the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
df.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc="std")
```

```python
# all cols together
df.pivot_table(index='sex', columns='smoker', values=['size','total_bill', 'tip'])
```

```python
# multidimenstional
df.pivot_table(index=['sex','smoker'],columns= ['day','time'],aggfunc={'size': 'mean','tip': 'max','total_bill': 'sum'},margins=True)
```

```python
# margins
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc="sum",margins=True)
```
