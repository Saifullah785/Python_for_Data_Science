```python
# importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplots

```python
# loading the dataset

iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# assigning numerical values to species column

iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
# ploting the scatter plot

plt.figure(figsize=(10,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c= iris['Species'],cmap='jet',alpha=0.9)
plt.title('Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

# **Annotations**

```python
# Loading another dataset, 'batter.csv', into a pandas DataFrame.
batter = pd.read_csv('batter.csv')
batter.head()
```

```python
# Taking a sample of 25 rows from the first 100 rows of the batter DataFrame.
# random_state ensures reproducibility of the sample.

sample_df = batter.head(100).sample(25,random_state=5)
```

```python
sample_df.sample(5)
```

```python
# Creating a figure with a specified size for the plot.

plt.figure(figsize=(18,10))

# Creating a scatter plot with 'avg' on the x-axis and 'strike_rate' on the y-axis.
# The size of the points is determined by the 'runs' column.

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

# Iterating through each row of the sample DataFrame to add annotations.

for i in range(sample_df.shape[0]):

  # Adding text annotations for each point.
  # The text is the value from the 'batter' column.
  # The position of the text is given by the 'avg' and 'strike_rate' values.

  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# **Horizontal and Vertical lines**

```python
# Defining lists for x and y coordinates.

x = [1,2,3,4]
y = [5,6,7,8]

# Creating a scatter plot of x and y.

plt.scatter(x,y)

# Adding text annotations to specific points on the plot.
# 'fontdict' is used to specify font properties like size and color.

plt.text(1,5,'point1',fontdict={'size':12, 'color': 'pink'})
plt.text(2,6,'point2',fontdict={'size':12, 'color': 'blue'})
plt.text(3,7,'point3',fontdict={'size':12, 'color': 'orange'})
plt.text(4,8,'point4',fontdict={'size':12, 'color': 'red'})
```

```python
plt.figure(figsize=(18,10))

# Creating a scatter plot with 'avg' on the x-axis and 'strike_rate' on the y-axis.
# The size of the points is determined by the 'runs' column.

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

# Adding horizontal lines to the plot at y=130 (red) and y=140 (green).

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

# Iterating through each row of the sample DataFrame to add annotations.

for i in range(sample_df.shape[0]):

   # Adding text annotations for each point, using values from the 'batter' column.

  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# **Subplots**

```python
# Displaying the first few rows of the 'batter' dataframe to see the data structure.
batter.head()
```

```python
# Creating a simple scatter plot of 'avg' vs 'strike_rate'.

plt.scatter(batter['avg'],batter['strike_rate'])
```

```python
# Creating a figure and a set of subplots.
# fig is the figure object, ax is an array of axes objects.
# In this case, we have one subplot.

fig, ax = plt.subplots(figsize=(15,10))
ax.scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
ax.set_title('Scatter Plot')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
batter.head()
```

```python
# Creating a figure and a grid of 2x1 subplots, sharing the x-axis.

fig, ax =plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,10))
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red',marker='D')
ax[1].scatter(batter['avg'],batter['runs'],color='blue',marker='+')

ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')

ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
ax[1].set_title('Average vs Runs')

fig.show()

```

```python
# Creating a figure and a grid of 2x2 subplots.

fig, ax =plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batter['avg'],batter['strike_rate'])
ax[0,1].scatter(batter['avg'],batter['runs'],color='blue',marker='+')
ax[1,0].hist(batter['avg'],color='green')
ax[1,1].hist(batter['runs'],color='orange')



fig.show()
```

```python
# Another way to create subplots using fig.add_subplot().
# Creating a figure.

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'])
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batter['avg'],batter['runs'],color='blue',marker='+')
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'],color='green')
ax4 = fig.add_subplot(2,2,4)
ax4.hist(batter['runs'],color='orange')



fig.show()
```

# **3D Surface Plots**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
```

```python
z.shape
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')
p=ax.plot_surface(x,y,z,cmap='viridis')
fig.colorbar(p)

ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

```python
z = np.sin(x) + np.cos(y)
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')
p=ax.plot_surface(x,y,z,cmap='viridis')
fig.colorbar(p)

ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

# **Contour Plots**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
```

```python


fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(x,y,z,cmap='viridis')

fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(x,y,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(x) + np.cos(y)
fig = plt.figure(figsize=(10,10))
ax = plt.subplot()
p=ax.contourf(x,y,z,cmap='viridis')
fig.colorbar(p)

ax.set_title('3D Surface Plot')
```
