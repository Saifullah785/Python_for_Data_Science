# **Bivariate Analysis**

categorical -> numerical and numerical ->numerical

Use case -> Time series data

```python
# plotting a simple function
price = [48000,54000,57000,4900,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(year,price)
plt.show()
```

```python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman.head()

```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
# labels title

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
# colors (hex) and line (width and style) and marker (size)

plt.plot(batsman['index'],batsman['V Kohli'],color= 'blue',linestyle= 'solid',linewidth=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color= 'red',linestyle='dotted') # dashdot,dashed
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#marker (size)
plt.plot(batsman['index'],batsman['V Kohli'],color= 'blue',linestyle= 'solid',linewidth=4 , marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color= 'red',linestyle='dotted') # dashdot,dashed
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color= 'blue',linestyle= 'solid',linewidth=4 , marker='D',markersize=10, label= 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color= 'red',linestyle='dotted',label='Rohit') # dashdot,dashed
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='upper left')
plt.show()
```

```python
# limiting axes

price = [48000,54000,57000,4900,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.xlim(2015,2021)
plt.ylim(0,75000)
plt.show()
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color= 'blue',linestyle= 'solid',linewidth=4 , marker='D',markersize=10, label= 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color= 'red',linestyle='dotted',label='Rohit') # dashdot,dashed
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='upper left')

plt.show() # show
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/content/batter.csv')
df=df.head(50)
df
```

```python
# size
tips = sns.load_dataset('tips')

tips.head()
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

# Bivariate Analysis

Numerical vs Categorical

Use case-Aggregate analysis of groups

```python
# simple bar chart
children = [10,30,40,10,20]
color = ['red','green','blue','yellow','pink']
plt.bar(color,children)
plt.show()
```

```python
# horizontal bar chart

plt.barh(color,children,color='purple')
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df.head()
```

```python
# Multiple Bar charts
plt.figure(figsize=(10,5))
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width=0.2)

plt.bar(np.arange(df.shape[0]) ,df['2016'],width=0.2)

plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.legend(['2015','2016','2017'])
plt.show()
```

```python
# xsticks
children = [10,30,40,10,20]
color = ['red red red red red','green green green green green','blue blue blue blue blue','yellow','pink']
plt.bar(color,children)
plt.xticks(rotation=45)
plt.show()
```

```python
# Stacked bar chart
plt.figure(figsize=(10,10))
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()
plt.show()
```
