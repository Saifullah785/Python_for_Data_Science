# **to_csv**

```python
# reads the 'deliveries.csv' file into a DataFrame and displays the first few rows.

df = pd.read_csv('deliveries.csv')
df.head()
```

```python
# groups the DataFrame by 'batsman', calculates the sum of 'batsman_runs', and resets the index.

temp_df =df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
# exports the temp_df DataFrame to a CSV file named 'batsman_score.csv', excluding the index.

temp_df.to_csv('batsman_score.csv',index=False)
```

```python
# creates a pivot table from the original DataFrame, with 'batsman' as the index,
# 'bowling_team' as columns, 'batsman_runs' as values, and 'sum' as the aggregation function.
# The resulting pivot table is then exported to a CSV file named 'batsman_vs_team.csv'.

df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

```python
# reads the 'batsman_vs_team.csv' file into a new DataFrame and displays the first few rows.

temp_df_1 = pd.read_csv('batsman_vs_team.csv')
temp_df_1.head()
```

```python
# This cell reads the 'batsman_score.csv' file into a DataFrame and displays it.

pd.read_csv('batsman_score.csv')
```

```python
# exports the temp_df DataFrame to an Excel file named 'batsman_score.xlsx', excluding the index.

temp_df.to_excel('batsman_score.xlsx',index=False)
```

```python
# This cell exports the temp_df DataFrame to an Excel file named 'batsman_score.xlsx',
# specifying the sheet name as 'batsman_score' and excluding the index.

temp_df.to_excel('batsman_score.xlsx',sheet_name='batsman_score',index=False)
```

```python
# This cell creates a pivot table from the original DataFrame, with 'batsman' as the index,
# 'bowling_team' as columns, 'batsman_runs' as values, and 'sum' as the aggregation function.

temp_df_2 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
# This cell demonstrates how to write multiple DataFrames to different sheets in a single Excel file.
# It uses a context manager to handle the ExcelWriter object.

with pd.ExcelWriter('batsman_vs_team.xlsx') as writer:
  temp_df.to_excel(writer,sheet_name='batsman_score_1',index=False)
  temp_df_2.to_excel(writer,sheet_name='batsman_score_2',index=False)
```

```python
# This cell reads the 'batsman_score.xlsx' file into a new DataFrame and displays the first few rows.

new_temp_df=pd.read_excel('batsman_score.xlsx')
new_temp_df.head()
```

```python

# This cell reads the 'batsman_vs_team.xlsx' file into a DataFrame,
# specifically reading the sheet named 'batsman_score_2', and displays the first few rows.

batsman_vs_team = pd.read_excel('batsman_vs_team.xlsx',sheet_name='batsman_score_2')
batsman_vs_team.head()
```

# **to_html**

```python
# This cell displays the first few rows of the original DataFrame.

df.head()
```

```python
# This cell filters the DataFrame to include only rows where 'batsman_runs' is 6,
# creates a pivot table with 'over' as index, 'ball' as columns, and counts the occurrences,
# and then exports the resulting pivot table to an HTML file named 'six_over.html'.


df.query('batsman_runs==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count').to_html('six_over.html')

```

```python
# This cell reads the 'six_over.html' file into a list of DataFrames and displays it.

pd.read_html('six_over.html')
```

# **To_Json**

```python
# This cell displays the first few rows of the original DataFrame

df.head()
```

```python
# This cell groups the DataFrame by 'batting_team' and 'batsman', calculates the sum of 'batsman_runs',
# reshapes the data with 'batting_team' as index and 'batsman' as columns, and exports the result to a JSON file named 'batsman_score.json'.

df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('batsman_score.json')
```

```python
# This cell reads the 'batsman_score.json' file into a new DataFrame and displays the first few rows.

df_02 = pd.read_json('batsman_score.json')

# This cell displays the first few rows of the original DataFrame.

df_02.head()
```

```python
# This cell imports the pymysql library and the create_engine function from sqlalchemy.

import pymysql
from sqlalchemy import create_engine
```

```python
## This cell shows how to create a SQLAlchemy engine for a MySQL database
## and export the original DataFrame to a SQL table named 'ipl_delivery'.
# # # {root}:{password}@{url}/{database}



engine = create_engine('mysql+pymysql://root:@localhost:3306/ipl')

df.to_sql('ipl_delivery', con = engine, if_exists = 'append')
```

```python
## This cell  shows how to export the temp_df DataFrame to a SQL table named 'batsman_runs'.

temp_df.to_sql('batsman_runs', con = engine, if_exists = 'append')
```
